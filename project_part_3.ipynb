## Introduction

The optic nerve sends visual information from the eye to the brain and has limited regenerative ability after damage. One way to promote regeneration is by targeting phosphatase and tensin homolog (PTEN). This study aimed to measure lipid concentrations in mice optic nerve crush samples following AAV-PLAP (control) or AAV-CRE (KO) injections to test if PTEN promotes regeneration.

observations. The lipid concentrations are the only data collected in this experiment.

regeneration after an injury. The hypothesis was that the deletion of PTEN promotes regeneration

concentrations for the control can be compared with the experimental group to determine if there 
are significant changes in lipids. If there are changes in the lipid concentrations, it can support 
that regeneration could be promoted by PTEN deletion.

### 3.1 Load and Clean Data

Downloading & Loading Data:

Here is a link to the article: https://www.ncbi.nlm.nih.gov/pmc/articles/PMC7797512/ 

At the bottom of the article, there is a supplementary section that includes the data that can be downloaded by clicking on the xlsx link.

Once the xlsx file is downloaded, the file can be opened in excel and downloaded as a csv file. The file was renamed 'lipids' and uploaded into JupyterLab.

```python
# import pandas 
import pandas as pd
# read in lipids data file
lipids = pd.read_csv("lipids.csv", index_col = 0)
```

```python
# display the first 5 rows of the dataframe
lipids.head(5)
```

To do the describe functions in 3.2, the data frame had to be transposed because without transposing the features would be the rows and the observations would be the columns.

```python
# import numpy 
import numpy as np
# transpose the dataframe
lipids = lipids.transpose()
# convert the transposed lipids into a dataframe
lipids = pd.DataFrame(lipids)
# print the first 5 rows of the transposed dataframe
lipids.head(5)
```

### 3.3 Visualizations

```python
# import seaborn
import seaborn as sns
sns.set_theme()
```
